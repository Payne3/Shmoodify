```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("billboard_lyrics_1964-2015.csv", encoding='ISO-8859-1')
df.head()
```

```python
#Shmoodify Wordcloud 

from wordcloud import WordCloud, STOPWORDS
plt.rcParams['figure.figsize']=(15,8.0)
plt.rcParams['font.size']=12                
plt.rcParams['savefig.dpi']=100              
plt.rcParams['figure.subplot.bottom']=.1 
stopwords = set(STOPWORDS)
data = df
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50, 
                          random_state=False
                         ).generate(str(data['Lyrics']))
#varibles for the wordcloud

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')

plt.show()

```
