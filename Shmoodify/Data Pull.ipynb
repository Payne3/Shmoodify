```python
#import dependencies

import pandas as pd
#from config import api_key 
import requests
import json
```

```python
#pull top rated songs after jan 1 2019

initial_url = "https://api.musixmatch.com/ws/1.1/track.search?apikey=a3400b624b6bf0d581420fff4d00fd3b&f_lyrics_language=en&f_track_release_group_first_release_date_min=20190101&page_size=100&page=100&s_track_rating=desc"
response = requests.get(initial_url)
musixmatch_data = response.json()
musixmatch_data
```

```python
musixmatch_data["message"]["body"]["track_list"][0]["track"]["primary_genres"]["music_genre_list"][0]["music_genre"]["music_genre_name"]
```

```python
#Put json response values into lists for the values we are interested in

trackid = []
track = []
rating = []
has_lyrics = []
instrumental = []
explicit = []
album = []
albumid = []
artist = []
artistid = []
genre = []

for song in musixmatch_data["message"]["body"]["track_list"]:
    trackid.append(song["track"]["track_id"])
    track.append(song["track"]["track_name"])
    rating.append(song["track"]["track_rating"])
    has_lyrics.append(song["track"]["has_lyrics"])
    instrumental.append(song["track"]["instrumental"])
    explicit.append(song["track"]["explicit"])
    album.append(song["track"]["album_name"])
    albumid.append(song["track"]["album_id"])
    artist.append(song["track"]["artist_name"])
    artistid.append(song["track"]["artist_id"])
    #genre.append(song["track"]["primary_genres"]["music_genre_list"][0]["music_genre"]["music_genre_name"])
```

```python
#Join all of the lists into a dataframe

song_df=pd.DataFrame()
song_df['trackid']= trackid
song_df['Song']= track
song_df['Rating']= rating
song_df['Lyrics']= has_lyrics
song_df['Instrumental']= instrumental
song_df['Explicit']= explicit
song_df['Album']= album
song_df['albumid']= albumid
song_df['Artist']= artist
song_df['artistid']= artistid
#song_df['Genre']= genre
song_df
```

```python
#Drop songs without available lyrics

song_df = song_df[song_df.Lyrics != 0]
song_df
```

```python
#pull the lyrics for each song in our dataframe

lyric_url="https://api.musixmatch.com/ws/1.1/track.lyrics.get"

lyrics = []

for index, row in song_df.iterrows():
    trackid = row['trackid']
    params = {
    "track_id": trackid,
    "apikey": "a3400b624b6bf0d581420fff4d00fd3b",
    }
    try:
        response = requests.get(lyric_url, params=params)
        lyrical_data = response.json()
        lyrics.append(lyrical_data["message"]["body"]["lyrics"]["lyrics_body"])
    except KeyError:
        lyrics.append(0)
song_df['Lyrics'] = lyrics
```
